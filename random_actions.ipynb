```python
import pandas as pd
import os
import random
from random import randint
```

```python
class RandomActions():
    def read_data(self,file='data/data.xlsx'):
        self.raw_data = pd.read_excel(file)
    def asign_data(self):
        should_list = []
        must_list = []
        for i in range(len(self.raw_data)):
            o = dict(self.raw_data.iloc[i])
            if o['must'] == 'y':
                must_list.append({'activity':o['activity'],'number':o['number'],'from':o['from']})
            if o['must'] == 'n':
                should_list.append({'activity':o['activity'],'number':o['number'],'from':o['from']})
        should_perc = self.raw_data['should_perc'].iloc[0]
        self.struct_data = {'should':should_list,'must':must_list,'should_perc':should_perc}
    def build(self):
        must_gg = []
        should_gg = []
        for i in self.struct_data['must']:
            interations = int(i['number']*(randint(i['from'],10)/10))
            if interations == 0:
                interations = 1
            act = [i['activity'] for k in range(interations)]
            must_gg = must_gg + act
        should_pr = random.choices(self.struct_data['should'],k=int(self.struct_data['should_perc']))
        for i in should_pr:
            act = [i['activity'] for k in range(int(i['number']*(randint(i['from'],10)/10)))]
            should_gg = should_gg + act
        all = should_gg+must_gg
        random.shuffle(all)
        return all
    def xlsx_result(self,how_many=1):
        result = {f"account{i}":self.build() for i in range(how_many)}
        max = 0
        for i in result:
            if len(result[i]) > max:
                max = len(result[i])
        for i in result:
            while True:
                if len(result[i]) < max:
                    result[i].append(' ')
                else:
                    break
        df = pd.DataFrame(result)
        df.to_excel('result.xlsx')
        return df
        
```

```python
c = RandomActions()
c.read_data('data/data.xlsx')
c.asign_data()
c.xlsx_result(5)
```
